# DataCamp Submission Draft

This notebook mirrors the final workbook-facing output from the local dbt + Python pipeline. The two DataFrames DataCamp expects are `group_predictions` and `knockout_predictions`.

## Methodology Summary

The submission is generated from a tested DuckDB/dbt feature warehouse and a Python tournament simulation layer. dbt cleans the raw fixtures, resolves stale playoff placeholders, builds point-in-time historical form features, joins FIFA ranking history, and creates team-level squad/event marts. Python trains expected-goals and direct-outcome models, blends exact-score probabilities with outcome probabilities, predicts corners/cards from event profiles, resolves the knockout bracket, and validates the final workbook schema.

Current holdout metrics: direct outcome accuracy `61.7%`, blended scoreline outcome accuracy `62.4%`, exact score accuracy `14.7%`, average goals MAE `0.907`.
